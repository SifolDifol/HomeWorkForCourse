1. Загрузите данные и выведите первые 10 строк DataFrame - ссылка - https://raw.githubusercontent.com/DataRepo2019/Data-files/master/subset-covid-data.csv.

```python
import pandas as pd
import numpy as np

data_frame_covid = pd.read_csv("https://raw.githubusercontent.com/DataRepo2019/Data-files/master/subset-covid-data.csv")
data_frame_covid.head(10)
```

2. Отфильтруйте строки DataFrame, где количество случаев больше 50.

```python
data_frame_covid.query('cases > 50')
```

3. Подсчитайте среднее количество случаев и смертей по каждому континенту.

```python
data_frame_covid.groupby('continent').agg({'deaths': 'mean' , 'cases': 'mean'})
```

4. Создайте новый столбец, который будет содержать количество случаев на миллион человек.

```python
data_frame_covid["countForOneMill"] = (data_frame_covid['cases'] / data_frame_covid['population']) * 1_000_000
data_frame_covid
```
