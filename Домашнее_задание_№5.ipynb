1. Взять датасет (приложенвматериалах - DZ_2_9.csv)
 в котором представлены цены на однокомнатные квартиры и возраст этих квартир. Проведите кластеризацию GMM, используя два признака: цену и возраст квартир. Какое оптимальное количество кластеров получилось ?

```python
gc = GaussianMixture(n_components=3)
pred = gc.fit_predict(home[['year','price']])
sns.scatterplot(x=home.year, y=home.price, c=pred)
```

2. Проведите кластеризацию с помощью алгоритма k-means. Какой алгоритм справился с задачей лучше?

```python
km = KMeans(n_clusters=3)
pred = km.fit_predict(home[['year', 'price']])
sns.scatterplot(x=home.year, y=home.price, c=pred)
```

3. Проведите кластеризацию DBSCAN с параметрами по умолчанию. Сколько точек было размечено как выбросы? Попробовать подобрать eps и min_samples так, чтобы ушел хотябы один выброс.

```python
dc = DBSCAN()
pred = dc.fit_predict(home[['year', 'price']])
sns.scatterplot(x=home.year, y=home.price, c=pred)
```

```python
dc = DBSCAN(
        eps = 1
        ,min_samples = 5
)
pred = dc.fit_predict(home[['year', 'price']])
sns.scatterplot(x=home.year, y=home.price, c=pred)
```
